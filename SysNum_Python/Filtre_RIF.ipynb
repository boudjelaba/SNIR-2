```python
from pylab import *
import scipy.signal as signal
from numpy import *
n = 61
a = signal.firwin(n, cutoff = 0.25, window = "hamming")

t = arange(0,2*pi,0.001)
b = random.randn(size(t))/10
x = sin(2*pi*t)+b
[m,p]=signal.freqz(a)

figure()
plot(abs(p))
title("Module")

figure()
plot(abs(m))
title("Phase")
y = signal.convolve(x,abs(p),mode='valid')

figure()
plot(x)
title("Signal bruité")

figure()
plot(y)
title("Signal bruité filtré")
```
