## Courbes des portes

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

n = np.arange(-21,22,1)
pr = []
for i in n:
    if np.abs(i)<=20:
        pr.append(1)
    else:
        pr.append(0)

phan = 0.5+0.5*np.cos((np.pi*n)/20)
pham = 0.54+0.46*np.cos((np.pi*n)/20)
pbl = 0.42+0.5*np.cos((np.pi*n)/20)+0.08*np.cos((2*np.pi*n)/20)
        
plt.figure(figsize=(6,4))
plt.plot(n,pr,label='Rectangulaire')
plt.plot(n,phan,label='Hanning')
plt.plot(n,pham,label='Hamming')
plt.plot(n,pbl,label='Blackman')
plt.grid()
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=2, mode="expand", borderaxespad=0.)
plt.show()
```
